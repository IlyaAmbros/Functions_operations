```python
## Data Structures - car database
# > functional programming
# > dict / list
# > JSON 
# > file
# > modular

import json
from random import randint
from faker import Faker
fake = Faker()

def loginprogram():
    nickname = input("Enter your name : ")
    return nickname

def createCar( model , brand , year , id , rent ):
    car = {}
    car["brand"] = brand
    car["model"] = model
    car["year"] = year
    car["id"] = id
    car["rent"] = rent
    return car

def printCarInfo(car):
    print ( "▬" * 50 )
    print ( f"{car['model']} : \n1) Release series - {car['brand']} ; \n2) Year of release - {car['year']} ; \n3) Identification code - {car['id']}." )
    print ( "▬" * 50 ) 
# 1. Show cars list
def printListCarInfo(cars):
    for c in cars :
        printCarInfo(c)
# 2. Add a car
def addCar(cars) :
    brand = input ( "Enter the brand of the new car : ")
    model = input ( "Enter the model of the new car : ")
    year = input ( "Enter the year of the new car : ")
    id = input ( "Enter the id of the new car : ")
    cars.append(createCar( model , brand , year , id ))
# 3. Edit a car
def editCar(cars) :
    car_id = input ( "Enter the id of the car to edit : ")
    car_index = -1
    for c in range ( len ( cars ) ) :
        if cars[c]["id"] == car_id :
            car_index = c
            break
    if car_index >= 0 :
        cars.pop( car_index )
        model = input ( "Enter a new model : ")
        brand = input ( "Enter a new brand : ")
        year = input ( "Enter a new year : ")
        id = input ( "Enter a new id : ")
        cars.insert( car_index , createCar( model , brand , year , id ))
# 4. Return a car
def removeCar( cars ):
    car_id = input ( "Enter the id of the car to edit : ")
    car_index = -1
    for c in range ( len ( cars ) ) :
        if cars[c]["id"] == car_id :
            car_index = c
            break
    if car_index >= 0 :
        cars.pop( car_index )
##################################################################################
# 2. Rent a car
def createClientList( cars , clientcars ) :
    for c in range ( len ( cars ) ) :
        if cars[c]["rent"] == True :
            clientcar = {}
            clientcar["client"] = fake.name()
            clientcar["id"] = cars[c]["id"]
            clientcar["birthdate"] = fake.date_of_birth()
            clientcar["city"] = fake.city()
            clientcar["number"] = fake.ssn()
            clientcars.append(clientcar)
    return clientcar

def printClientRentCarMenu( cars , clientcars ) :
    print("▩"*50,"\n")
    for c in range ( len ( cars ) ) :
        if cars[c]["rent"] == False :
            print (f"▶ A car with an identification code | {cars[c]['id']} | is not rented by anyone .\n" )
    print("List of rented cars :\n")
    for i in range ( len ( clientcars ) ) :
        print(f"▶ A car with an identification code | {clientcars[i]['id']} | is rented by a {clientcars[i]['client']} : \n1) Date of birth - {clientcars[i]['birthdate']} ; \n2) Place of residence - {clientcars[i]['city']} ; \n3) Phone number - +{clientcars[i]['number']} .\n")
    print("▩"*50)

def rentCar( cars , clientcars , nickname ) :  
    car_rent = input ("Enter the identification code of the car you want to rent : ")
    for c in range ( len ( cars ) ) :
        if car_rent == cars[c]["id"] :
            if cars[c]["rent"] == True :
                print("You cannot rent this car because it is already rented !")
            else :
                ("To rent a car, you need to fill out a form : ")
                client_name = nickname
                client_birthdate = input("2) Enter your date of birth : ")
                client_city = input("3) Enter your place of residence : ")
                client_phone = int(input("4) Enter your phone number : +"))
                print ("\nRent a car for one day is $ 50 .\nThe maximum rental time is 90 days .\n")
                arenda_duration = int(input("Enter the number of days you are renting a car for : "))
                if 0 <= arenda_duration > 90 :
                    print ("Such a client already exists !")
                else :
                    print (f"The cost of your rental is : {arenda_duration * 50} $ . ")
                    for i in range ( len ( clientcars ) ) :
                        if client_name not in clientcars[i]["client"] :
                            clientcar = {}
                            clientcar["client"] = client_name
                            clientcar["id"] = cars[c]["id"]
                            clientcar["birthdate"] = client_birthdate
                            clientcar["city"] = client_city
                            clientcar["number"] = client_phone
                            clientcars.append( clientcar )
                            cars[c]["rent"] = True
    return clientcars
# 3. Return a car
def returnCar( cars , clientcars , nickname ) :
    client_index = -1
    car_return = input ("Enter the identification code of the car you want to return : ")
    for c in range ( len ( clientcars ) ) :
        if car_return == clientcars[c]["id"] and nickname == clientcars[c]["client"] :
            client_index = c
    if client_index >= 0 :
        for i in range ( len ( cars ) ) :
            if cars[i]["id"] == car_return :
                    cars[i]["rent"] = False
        clientcars.pop( client_index )
    return clientcars

# 3. Save Data
def saveCarData(cars) :
    file = open("data/cars.json" , "w")
    file.write( json.dumps(cars))
    file.close()

def loadCarData(cars) :
    file = open("data/cars.json" , "r")
    cars = json.loads(file.read)
    file.close()
    return cars
```
